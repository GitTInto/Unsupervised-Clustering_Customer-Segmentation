```python
#Loading the dataset
data = pd.read_csv("./marketing_campaign.csv", sep="\t")
print("Number of records:", len(data))
data.head()
```

```python
#Information on features 
data.info()
```

```python
#There are missing values in income
#To remove the NA values
data = data.dropna()
print("The  number of records after dropping missing values:", len(data))
```

```python
## Feature Engineering, Data Cleaning, Encoding
```

```python
### Dt_Customer to Customer_For
```

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"])
```

```python
dates = []
for i in data["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
```

```python
#Dates of the newest and oldest recorded customer
print("The newest customer's date:",max(dates))
print("The oldest customer's date:",min(dates))
```

```python
#Created a feature "Customer_For"
days = []
d1 = max(dates) #taking it to be the newest customer
for i in dates:
    delta = d1 - i
    days.append(delta)
data["Customer_For"] = days
data["Customer_For"] = pd.to_numeric(data["Customer_For"], errors="coerce")
```

```python
data["Customer_For"]
```

```python
### Marital_Status
```

```python
print("Types Marital_Status:\n", data["Marital_Status"].value_counts())
```

```python
#Deriving living with by marital status

#Living with Partner = 1
#living alone = 0

data["Living_With"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})
#data["Living_With"]=data["Marital_Status"].replace({"Married":1, "Together":1, "Absurd":0, "Widow":0, "YOLO":0, "Divorced":0, "Single":0,})
```

```python
### Year_Birth to Age
```

```python
data["Age"] = 2021-data["Year_Birth"]
```

```python
### Children
```

```python
data["Children"]=data["Kidhome"]+data["Teenhome"]
```

```python
### Is parent?
```

```python
data["Is_Parent"] = np.where(data.Children> 0, 1, 0)
```

```python
### Total spendings
```

```python
data["Spent"] = data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]
```

```python
#Renaming some columns
data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
### Members in household
```

```python
data["Family_Size"] = data["Living_With"].replace({"Alone": 1, "Partner":2})+ data["Children"]
```

```python
### Education
```

```python
print("Types of Education:\n", data["Education"].value_counts())
```

```python
#Undergraduate = 0
#Graduate = 1


data["Education"]=data["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
```

```python
#Dropping the redundant features
to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
data = data.drop(to_drop, axis=1)
```

```python
data.describe()
```

```python
data.info()
```

```python
### Label Encoding
```

```python
# list the categorical variables
s = (data.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables are:", object_cols)
```

```python
#Label Encoding the object dtypes.
LE=LabelEncoder()
for i in object_cols:
    data[i]=data[[i]].apply(LE.fit_transform)
```

```python
#To plot some selected features 
#Setting up colors prefrences
#sns.set(rc={"axes.facecolor":"#FFF9ED","figure.facecolor":"#FFF9ED"})
#pallet = ["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"]
#cmap = colors.ListedColormap(["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"])
#Plotting following features
To_Plot = [ "Income", "Recency", "Customer_For", "Age", "Spent", "Is_Parent"]
print("Reletive Plot Of Some Selected Features: A Data Subset")
plt.figure()
sns.pairplot(data[To_Plot], hue= "Is_Parent",palette= (["#FF0000","#0000FF"]))
#Taking hue 
plt.show()
```

```python
#Dropping the outliers by setting a cap on Age and income. 
data = data[(data["Age"]<90)]
data = data[(data["Income"]<600000)]
print("The total number records after dropping the outliers:", len(data))
```

```python
### Removing the un-necessary collumns
```

```python
ds = data.copy()
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis=1)
```

```python
### Standard Scaling
```

```python
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
scaled_ds.head()
```

```python
scaled_ds.info()
```

```python
#correlation matrix
corrmat= scaled_ds.corr()
plt.figure(figsize=(20,20))  
sns.heatmap(corrmat,annot=True,  center=0)
```

```python
## Dimentionality Reduction
```

Many of these features are correlated, and hence redundant. This is why I will be performing dimensionality reduction on the selected features before putting them through a classifier.
Dimensionality reduction is the process of reducing the number of random variables under consideration, by obtaining a set of principal variables.

Principal component analysis (PCA) is a technique for reducing the dimensionality of such datasets, increasing interpretability but at the same time minimizing information loss.

```python
# PCA to reduce dimentions (features) to 3
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
# 3D plot Of data after the dimention reduction
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="blue", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

```python
## Clustering
```

```python
### Elbow method to determine the apt number of clusters
```

```python
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

Elbow method shows us that four is the optimal number of clusters.

```python
### Agglomerative Clustering
```

```python
AC = AgglomerativeClustering(n_clusters=4)
yhat_AC = AC.fit_predict(PCA_ds)
```

```python
PCA_ds["Clusters"] = yhat_AC

data["Clusters"]= yhat_AC
```

```python
#Plotting the clusters
cmap = colors.ListedColormap(["#FF0000","#0000FF","#00FF00","#001308"])
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = cmap )
ax.set_title("Data in Clusters")
plt.show()
```
